```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
girls_names = []

for i in range(1980, 2023):
    url = f'https://www.ssa.gov/oact/babynames/state/top5_{i}.html'
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    
    tables = soup.find_all("table")
    
    
    if len(tables) >= 2:
        table = tables[0]
        
        rows = table.find_all("tr")
        
        for row in rows[1:]:  
            columns = row.find_all("td")
            
            if len(columns) >= 6:
                state = columns[0].text.strip()
                rank_1 = columns[1].text.strip()
                rank_2 = columns[2].text.strip()
                rank_3 = columns[3].text.strip()
                rank_4 = columns[4].text.strip()
                rank_5 = columns[5].text.strip()
                
                data = {
                    "year": i,
                    "state": state,
                    "Rank 1": rank_1,
                    "Rank 2": rank_2,
                    "Rank 3": rank_3,
                    "Rank 4": rank_4,
                    "Rank 5": rank_5
                }
                
                girls_names.append(data)
```

```python
girls_names
```

```python
girls_df = pd.DataFrame(girls_names)

girls_df.head(50)
```

```python
girls_df.to_csv("data/girls_data.csv")
```
